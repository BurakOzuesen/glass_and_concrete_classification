```python
import pandas as pd
import random

data = pd.read_csv("glass.csv")


def cross_validation(data):

    INPUT_SIZE = data.shape[0]
    indices = []

    for i in range(INPUT_SIZE):         # np.arange fonksiyonu da kullanılabilir
        indices.append(i)

    random.shuffle(indices)

    bucket_1 = []
    bucket_2 = []
    bucket_3 = []
    bucket_4 = []
    bucket_5 = []

    for index, item in enumerate(indices):          # 5erli sub arrayler olarak da ele alınabilir ama galiba en hızlısı bu
        if index % 5 == 0:
            bucket_1.append(item)
        elif index % 5 == 1:
            bucket_2.append(item)
        elif index % 5 == 2:
            bucket_3.append(item)
        elif index % 5 == 3:
            bucket_4.append(item)
        else:
            bucket_5.append(item)
    
    data_subset_1 = data.iloc[bucket_1,:]
    data_subset_2 = data.iloc[bucket_2,:]
    data_subset_3 = data.iloc[bucket_3,:]
    data_subset_4 = data.iloc[bucket_4,:]
    data_subset_5 = data.iloc[bucket_5,:]

    return data_subset_1, data_subset_2, data_subset_3, data_subset_4, data_subset_5


def normalization(data):
    for column_name in data.columns:
        
        if column_name == "Type":
            continue

        column_max = data[column_name].max()
        column_min = data[column_name].min()

        for i in range(len(data[column_name])):
            
            data[column_name][i] = (data[column_name][i] - column_min) / (column_max - column_min)
            
    return data
```

```python
data_subset_1, data_subset_2, data_subset_3, data_subset_4, data_subset_5 = cross_validation(data)
```
